### 양육비 데이터 분석

```python
import pandas as pd
```

```python
df2021=pd.read_excel('./data/양육비.xlsx',index_col=0,sheet_name='2021')
df2021
```

```python
df2021['2021평균']=df2021.mean(axis=1)
df2021['2021 1년치']=df2021['2021평균']*12
df2021
```

```python
df2019=pd.read_excel('./data/양육비.xlsx',index_col=0,sheet_name='2019')
df2019
```

```python
df2019['2019평균']=df2019.mean(axis=1)
df2019['2019 1년치']=df2019['2019평균']*12
df2019
```

```python
df=pd.DataFrame([df2019['2019 1년치'],df2021['2021 1년치']]).T
df
```

```python
df['증감률'] = abs((df['2021 1년치'] - df['2019 1년치'])*100/df['2019 1년치'])/2
df
```

```python
for i in range(2022, 2053):
    if i == 2022:
        df[i] = df['2021 1년치'] * (1+(df['증감률']/100))

    else:
        df[i] = df[i-1] * (1+(df['증감률']/100))
df
```

```python
df.drop(columns=['2019 1년치','2021 1년치','증감률',2022],axis=1,inplace=True)
```
