```python
import pandas as pd
file=r'cow_data.csv'
df=pd.read_csv(file)
df
```

```python
df_=df.iloc[:,:-1].copy()
```

```python
df_.drop(index=df_[df_['최저가']==0].index)
```

```python
# 낙찰가에 이상치가 있어서 전처리한다
df_['낙찰가']=df_['낙찰가'].apply(lambda x: int(x/10) if x>2000 else int(x))
df_
```

```python
df_.성별.replace('암',0,inplace=True)
df_.성별.replace('수',1,inplace=True)
df_.종류.replace('큰소',0,inplace=True)
df_.종류.replace('혈통우',1,inplace=True)
df_.종류.replace('일반우',2,inplace=True)
df_

```

```python
df_data=df_[['성별','중량','최저가','낙찰가','결격사유 여부','종류']]
df_data
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
# 가정: wqDF_nona가 DataFrame 형태의 데이터입니다.
sns.pairplot(data=df_data,hue='종류')
plt.show()

plt.tight_layout()
plt.show()
```

```python
# 상관계수 높은 것만 df에 나타남
def corr_name(df):
    df_name=df.corr(numeric_only=True).applymap(lambda x: x if (x >= 0.7) else (x if (x <= -0.7) else ''))
    return df_name

corr_name(df_data)
```

```python
# 낙찰가가 범주형인 종류에 영향을 받는 것을 볼 수 있으므로 stratify=df_data['종류']
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df_data, test_size=0.2,random_state=0,stratify=df_data['종류'])
```
