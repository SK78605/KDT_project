## 스마트폰 일 평균 이용시간

```python
import pandas as pd
import numpy as np
```

## [1] 스마트폰 1회 이용 시 평균 이용시간

```python
# 국가통계포털 kosis
# 스마트폰 1회 이용 시 평균 이용시간 파일
timefile='data/스마트폰_1회_이용_시_평균_이용시간_20230713192823.xlsx'
time=pd.read_excel(timefile,index_col=[0,1,2],header=[0,1,2])
time.head()
```

```python
# 연령대별만 남김
time=time.iloc[5:17]
# 과의존위험군만 남김
time=time.iloc[1::3]
```

```python
# 평균 열만 남김
a=[]
for n in range(time.shape[1]):
    if time.columns.to_list()[n][2]=='평균 (분)':
        a.append(n)
```

```python
time=time.iloc[:,a]
time
```

```python
# 연령대 별 1회 평균 이용시간 열 추가
time['1회 평균 이용시간']=round(time.mean(axis=1),1).to_list()
time
```

## [2] 스마트폰 1일 평균횟수

```python
# 스마트폰 1일 평균횟수
countfile='data/스마트폰_1일평균_이용횟수_20230713192909.xlsx'
countdf=pd.read_excel(countfile,index_col=[0,1,2],header=2)
countdf.head()
```

```python
# 연령대별만 남김
countdf=countdf.iloc[5:17]
# 과의존위험군만 남김
countdf=countdf.iloc[1::3]
```

```python
# 평균 열만 남김
countdf=countdf.iloc[:,4::5]
```

```python
# 연령대 별 1일 평균 이용 횟수
countdf['1일 평균 이용 횟수']=round(countdf.mean(axis=1),2).to_list()
countdf
```
