```python
import pandas as pd
import numpy as np
```

```python
#파일 불러오기
filePisa='문해력파일/학업_성취도_OECD회원국__20230714022025.xlsx'
pisa=pd.read_excel(filePisa)
pisa.head(3)
```

```python
pisa.info()
```

```python
# 결측 값 제거
pisa.dropna(how='any',inplace=True)
pisa=pisa.T
pisa.head()
```

```python
pisa.columns = list(pisa.iloc[0])
pisa.drop('국가별', axis=0, inplace=True)
pisa.head()
```

```python
# 읽기점수만 남기기
dates=pisa['국가별'].str.split('-')
mask= dates.str.get(0) =='읽기점수'
pisa=pisa[mask]
pisa
```

```python
# 날짜 timestamp 변환 후 년도만 남겨서 열 추가
pisa['new_date']=pd.to_datetime(pisa.index)
pisa['new_date']=pisa['new_date'].dt.year
pisa.head()
```

```python
# 열 이름 빈칸 없애기
pisa.columns=[str(n).strip() for n in pisa.columns.to_list()]
pisa
```

```python
pisaDF=pisa[["new_date","국가별","대한민국"]].groupby(by='new_date').mean()
pisaDF
```

```python
pisaDF=pisaDF.T
pisaDF
    
```
