```python
import fontda
import pandas as pd
import numpy as np
```

```python
file2='문해력파일/주로_이용하는_SNS_계정_123순위_20230713192732.xlsx'
df22=pd.read_excel(file2,header=1,index_col=1)
df22.head()
```

```python
df22=df22.iloc[4:9,2:13]
df22
```

```python
df22=df22.drop(['싸이월드미니홈피 (%)','빙글 (%)','비트윈 (%)','웨이보 (%)','핀터레스트 (%)'],axis=1)
df22
```

```python
df22=df22[['트위터(Twitter) (%)','페이스북(Facebook) (%)','카카오스토리 (%)','네이버밴드 (%)','인스타그램(instagram) (%)','틱톡 (%)']]
df22.columns=['트위터','페이스북','카카오스토리','네이버밴드','인스타그램','틱톡']
df22
```

```python
df22=df22.iloc[[0]]
df22
```

```python
df22['sum']=df22.sum(axis=1)
df22
```

```python
df22=df22.div(df22['sum'],axis=0)
df22
```

```python
df22=df22.apply(lambda x: round(x*100,1),axis=0)
df22
```

```python
df22.drop('sum',axis=1,inplace=True)
```
