# [1] 헷갈리는 맞춤법

```python
# 출처 : 네이버 검색어 트렌드 데이터랩
```

```python
import pandas as pd
import numpy as np
```

```python
# 네이버 검색어 트렌드
file='헷갈리는맞춤법.xlsx'
hardWord=pd.read_excel(file,header=6,index_col=0)
hardWord.head()
```

```python
# 날짜 컬럼 제거

hardWord=hardWord.drop(['날짜.1','날짜.2','날짜.3','날짜.4','날짜.5','날짜.6','날짜.7'],axis=1)
hardWord.head()
```

```python
hardWord.info()
```

```python
# nan 값 확인  => 없음
hardWord.isnull().sum()
```

```python
# 년도별 합계 리스트에 담기
sum16=round(hardWord.iloc[:12].sum()).to_list()
sum17=round(hardWord.iloc[12:24].sum()).to_list()
sum18=round(hardWord.iloc[24:36].sum()).to_list()
sum19=round(hardWord.iloc[36:48].sum()).to_list()
sum20=round(hardWord.iloc[48:60].sum()).to_list()
sum21=round(hardWord.iloc[60:72].sum()).to_list()
sum22=round(hardWord.iloc[72:84].sum()).to_list()
```

```python
# 년도 합계 데이터프레임 생성
hard=pd.DataFrame([sum16,sum17,sum18,sum19,sum20,sum21,sum22],index=['2016','2017','2018','2019','2020','2021','2022']
                  ,columns=['되 돼','데 대','던 든','어떡해 어떻게','안 않','낳다 낫다','왠 웬','율 률'])
hard["-"]=" "
# 합계 열 추가
hard['합계']=hard.sum(axis=1,numeric_only=True)
```

```python
# 퍼센트 열 추가
total=hard['합계'].sum()
a=[round(n/total*100,2)  for n in hard['합계'].to_list()]
hard['퍼센트']=a
```

```python
# 증가값 열 추가
hard1=hard.iloc[:,-1:]
hard2=hard1.copy()
hard2=hard2.rename(index={str(int(x)):str(int(x)+1)  for x in hard1.index})
```

```python
ret=hard1-hard2
ret.replace(np.nan,0,inplace=True)
ret.drop(index=['2023'],inplace=True)
```

```python
hard['증가값']=ret.values
```

## [2] 국제 학업 성취도 평가 통계

```python
file='학업_성취도_OECD회원국__20230714022025.xlsx'
pisa=pd.read_excel(file,header=[0,1])
pisa=pisa.set_index([("국가별","국가별")])   # set_index()
pisa.head()
```

```python
# 읽기점수 열만 남기기
pisa=pisa[[('2012', '읽기점수-남자'),('2012', '읽기점수-여자'),('2015', '읽기점수-남자'), ('2015', '읽기점수-여자'),('2018', '읽기점수-남자'),
 ('2018', '읽기점수-여자')]]
```

```python
pisa.info()
```

```python
# 결측치 체크   --> 있음
pisa.isnull().sum()
```

```python
ret=[]
for n in range(pisa.shape[0]):ret.extend(pisa.iloc[n].unique())
a=[]
for m in ret:
    if not isinstance(m,int): a.append(m) 
a
```

```python
# 결측치 처리
pisa.replace('-',np.nan,inplace=True)
pisa=pisa.dropna(how='all')
pisa.head()
```

```python
# 한국만 남기기
pisa=pisa.iloc[0:1]
# 년도별 평균 내서 열 추가
mean12=pisa.iloc[:,:2].mean(axis=1).values
mean15=pisa.iloc[:,2:4].mean(axis=1).values
mean18=pisa.iloc[:,4:].mean(axis=1).values
pisa['2012년']=mean12
pisa['2015년']=mean15
pisa['2018년']=mean18
pisa=pisa.iloc[:,6:]
pisa=pisa.reset_index(drop=True)    #reset_index(drop=True)
pisa
```

```python
# 증감 열 추가
pisa1=pisa.copy()
pisa1.columns=[0,1,2]
pisa2 = pisa1.copy()
pisa2 = pisa2.rename(columns={x:x+1 for x in pisa1.columns})
```

```python
dec=pisa1-pisa2
dec.drop(3,axis=1,inplace=True)
dec.replace(np.nan,0,inplace=True)
```

```python
pisa.loc[' ']=" "
pisa.loc['증감']=list(dec.loc[0])
pisa.rename(index={0:"대한민국"},inplace=True)
```

### 결론 :  학생들의 읽기능력이 저하되고 있음
